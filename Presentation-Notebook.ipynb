## Health Submetric
To represent a neighborhood's health, the median age at death was chosen. In other words, the neighborhoods were ranked by lifespan in years. The higher the value, the longer the median lifespan per community. Lifespan was chosen to measure healthiness as it provides a general overview into any person's health. The healthier you are, the longer you typically live. The dataset uses information from 2011-2015 and is based on the individual's residence location, not on the location of the death. The information was collected by the Pennsylvania Department of Health.

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Reading in the CSV file
death = pd.read_csv("life.csv")

# Sorting the total median age at death by neighborhood, in descending order
sorteddeath = death.sort_values(by='TOTAL MD AGE AT DEATH', ascending=False)

# Giving it some style :)
plt.style.use('seaborn')
# Creating and formatting the plot
deathplot = sorteddeath.head(15).plot(fontsize = '13',
                                    kind='bar', 
                                    x='NEIGHBORHOOD',
                                    y='TOTAL MD AGE AT DEATH',
                                    rot = 90,
                                    ylabel='Median Age at Death (Years)', 
                                    xlabel='Neighborhood', 
                                    title='Top 10 Neighborhoods by Median Age at Death', 
                                    figsize=(10,10))
plt.ylim(70,90)
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
povertyFile= pd.read_csv('poverty.csv')#import data
poverty = povertyFile.loc[:,['Neighborhood','Total:','below poverty level:']] #filter out useless data
poverty['Proportion Below poverty line'] = poverty['below poverty level:']/poverty['Total:'] # new column "% below poverty line"
poverty = poverty.drop([71], axis=0) #delete SouthShore bc not enough data to be useful
poverty = poverty.sort_values('Proportion Below poverty line', ascending=False) #sort by heighest to lowest % poverty
plt.style.use('classic') 
plt.rcParams["figure.figsize"] = (20, 10) #no idea how this works but it makes it bigger
poverty.plot(x='Neighborhood',y='Proportion Below poverty line',kind='bar')
plt.title('Poverty by Neighborhood')
plt.ylabel('% below poverty line')
```
