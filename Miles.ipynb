## Intro:
For my notebook, I will be looking at poverty rates in Pittsburgh neighborhoods. A lower poverty rate indicates a wealthier, happier, and overall better neighborhood. For my data, I found a data set that gave a population and number of people living in poverty. I then found the poverty rate in each neighborhood and created a bar graph to visualize how the neighborhoods related to eachother.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
povertyFile= pd.read_csv('poverty.csv')#import data
poverty = povertyFile.loc[:,['Neighborhood','Total:','below poverty level:']] #filter out useless data
poverty['Proportion Below poverty line'] = poverty['below poverty level:']/poverty['Total:'] # new column "% below poverty line"
poverty = poverty.drop([71], axis=0) #delete SouthShore bc not enough data to be useful
poverty = poverty.sort_values('Proportion Below poverty line', ascending=False) #sort by heighest to lowest % poverty
plt.style.use('classic') 
plt.rcParams["figure.figsize"] = (20, 10) #no idea how this works but it makes it bigger
poverty.plot(x='Neighborhood',y='Proportion Below poverty line',kind='bar')
plt.title('Poverty by Neighborhood')
plt.ylabel('% below poverty line')
```

```python
plt.rcParams["figure.figsize"] = (10, 5)#smaller
poverty[:10].plot.bar(x='Neighborhood', y='Proportion Below poverty line')#plot first 10
plt.xticks(rotation=70)#rotate x axis
plt.title('10 Neighborhoods with the HEIGHEST Poverty Rates')
plt.ylabel('% below poverty line')
```

```python
povertyBottom = poverty.sort_values('Proportion Below poverty line', ascending=True)#sort lowest to heighest
povertyBottom[:10].plot.bar(x='Neighborhood', y='Proportion Below poverty line')#plot first 10
plt.xticks(rotation=70)#rotate x axis
plt.title('10 Neighborhoods with the LOWEST Poverty Rates')
plt.ylabel('% below poverty line')
```

```python
plt.rcParams["figure.figsize"] = (15, 10)#bigger
povertyBottom[:40].plot.bar(x='Neighborhood', y='Proportion Below poverty line')#lowest 40
plt.xticks(rotation=90)#rotate x axis
plt.title('40 Lowest Poverty Rates')
plt.ylabel('% below poverty line')
```
